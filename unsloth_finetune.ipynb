```python
# Load model with 4-bit quantization <button class="citation-flag" data-index="3">
from unsloth import FastLanguageModel

max_seq_length = 2048
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="unsloth/DeepSeek-R1-Distill-Llama-8B",  # Verified identifier <button class="citation-flag" data-index="3">
    max_seq_length=max_seq_length,
    dtype=None,
    load_in_4bit=True,
    token=hf_token,
)

# Simplified DSA prompt template (no explanations) <button class="citation-flag" data-index="6">
prompt_template = """Solve this Data Structures & Algorithms problem:

### Question:
{}

### Final Answer:
{}
"""

EOS_TOKEN = tokenizer.eos_token  # Required for proper formatting
```

```python
def format_dsa_examples(examples):
    """Format DSA Q/A pairs without explanations"""
    questions = examples["question"]
    answers = examples["answer"]
    formatted = []
    for q, a in zip(questions, answers):
        formatted.append(prompt_template.format(q, a) + EOS_TOKEN)
    return {"text": formatted}

# Load and preprocess your DSA dataset
from datasets import load_dataset

dataset = load_dataset("json", data_files="dsa_data.json")["train"]
dataset = dataset.map(
    format_dsa_examples,
    batched=True,
    remove_columns=["question", "answer"]  # Remove original columns <button class="citation-flag" data-index="7">
)

# Apply LoRA configuration optimized for reasoning tasks <button class="citation-flag" data-index="9">
model = FastLanguageModel.get_peft_model(
    model,
    r=16,
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj",
                   "gate_proj", "up_proj", "down_proj"],
    lora_alpha=16,
    lora_dropout=0.05,
    bias="none",
    use_gradient_checkpointing="unsloth",  # Memory optimization <button class="citation-flag" data-index="4">
    random_state=3407,
)
```

```python

# Training configuration with knowledge-informed parameters <button class="citation-flag" data-index="5"><button class="citation-flag" data-index="6">
from trl import SFTTrainer
from transformers import TrainingArguments

trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=dataset,
    dataset_text_field="text",
    max_seq_length=max_seq_length,
    args=TrainingArguments(
        per_device_train_batch_size=4,  # Adjust based on GPU memory
        gradient_accumulation_steps=2,
        warmup_steps=50,
        num_train_epochs=5,  # Optimal for 700 samples <button class="citation-flag" data-index="5">
        learning_rate=2e-4,
        fp16=True,
        logging_steps=10,
        optim="adamw_8bit",  # 8-bit optimizer from Unsloth <button class="citation-flag" data-index="1">
        weight_decay=0.01,
        lr_scheduler_type="cosine",
        seed=3407,
        output_dir="./dsa-finetuning-results",
    ),
)

```

```python
# Start training with Unsloth optimizations <button class="citation-flag" data-index="3">
trainer.train()

# Save final model (merged for easy deployment)
model.save_pretrained_merged(
    "DeepSeek-R1-DSA-Finetuned",
    tokenizer,
    save_method="merged_16bit"
)
```
